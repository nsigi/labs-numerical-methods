# Создание матрицы. Работа с элементами матрицы

1. Создать квадратную матрицу из случайных целых чисел из [0,8] размера 6 . Создать две новые матрицы: первая – из двух последних строк исходной матрицы (должна получиться матрица размера 2х6), вторая – из двух первых столбцов матрицы (матрица размера 6х2).

```python
A = np.random.randint(0, 9, (6, 6))
print(A)
B = A[-2:]  # срез последних 2-х строк
B
```

```python
C = A[:, :2]  # срез первых 2-х строк
C
```

2. Выполнить умножение матриц из предыдущего пункта тремя способами: 
1) используя векторный алгоритм умножения матриц 

```python
D = np.zeros((2, 2), dtype=int)
for i in range(2):
    for j in range(2):
        for p in range(6):
            # пользуемся правилом перемножения матриц
            D[i, j] += B[i, p]*C[p, j]
```

2) используя матричный алгоритм, записав матрицу С по строкам 

```python
F = np.dot(B, C)
for i in range(2):
    F[i, :] = np.dot(B[i, :], C)
```

3) проверив с помощью функции np.dot.

```python
G = np.zeros((2, 2), dtype=int)
for i in range(2):
    for j in range(2):
        G[i, j] = np.dot(B[i, :], C[:, j])
```

```python
print('Векторный алгоритм\n', D)
print('Матричный алгоритм\n', F)
print('np.dot\n', G)
```

# Решение системы линейных уравнений с треугольной матрицей А

3. Создать произвольную верхнетреугольную матрицу А 4 порядка (не унитреугольную), вектор B произвольный. Решить систему AX = B.

```python
# генерируем  верхнетреугольную (не унитреугольную) матрицу А 4-го порядка
A = np.triu(np.random.randint(-5, 6, (4, 4)))
B = np.array([6, 2, 4, -1], int)
x = np.zeros((4, 1), float)

x[3] = B[3]/A[3, 3]

for i in range(2, -1, -1):
    x[i] = (B[i] - np.dot(A[i, i+1:], x[i+1:]))/A[i, i]

print(A)
print(B)
print(x)
```

```python
# проверка решения
np.linalg.solve(A, B)
```

# LU разложение матрицы А

4. Решить систему, используя LU разложение 

```python
A = np.array([[15.7, 6.6, -5.7, 11.5], [8.8, -6.7, 5.5, -4.5],
              [6.3, -5.7, -23.4, 6.6], [14.3, 8.7, -15.7, -5.8]], float)
B = np.array([-2.4, 5.6, 7.7, 23.4], float)
A
```

```python
L = np.identity(4, float)  # нижняя унитреугольная матрица
U = np.zeros((4, 4), float)  # верхняя треугольная матрица
for i in range(4):
    for j in range(4):
        if i > j:
            L[i, j] = (A[i, j] - np.dot(L[i, :j], U[:j, j])) / U[j, j]
        else:
            U[i, j] = A[i, j] - np.dot(L[i, :i], U[:i, j])
```

```python
Y = np.array([0, 0, 0, 0], float)
Y[0] = B[0]

for i in range(1,4):
    Y[i] = B[i] - np.dot(L[i, :i], Y[ :i])
print(np.dot(L, Y) == B)
```

```python
x[3] = Y[3]/U[3, 3]
for i in range(2, -1, -1):
    x[i] = (Y[i] - np.dot(U[i, i+1:], x[i+1:]))/U[i, i]
# проверка решения
Ch = np.linalg.solve(A, B)
print(Ch, '\n')
for i in range(4):
    print(i + 1, ':', x[i])
```

```python
#дргуой спосооб нахождения Y
Y = np.array([0, 0, 0, 0], float)
Y[0] = B[0]  # находим сразу x1, зная, что L - нижняя унитреугольная матрица
for i in range(1, 4):
    s = 0  # инициализация для переменной, хранящей значение левой части
    for j in range(i):
        # вычисляем выражение в левой части на основе уже известных x
        s += L[i, j] * Y[j]
    Y[i] = B[i] - s  # переносим всё из левой части в правую кроме неизвестного х

print(np.dot(L, Y) == B)
```
